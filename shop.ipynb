```python
data.isnull().sum()#checking existence of none values
```

```python
product_line = data['Product line'].unique()
product_line
```

```python
payment_type = data['Payment'].unique()
payment_type
```

```python
data['Datetime'] = data['Date']+' '+data['Time']
data['Datetime'] = data['Datetime'].apply(pd.to_datetime)#casting object to datetime
data = data.drop(['Date','Time'],axis=1).sort_values(by='Datetime',ignore_index=True)
data
```

```python
data.corr()
```

```python
data.info()
```

```python
branch_a = data[data['Branch'] == 'A'].reset_index(drop = True).drop('Branch',axis = 1)#dataframe of 'A Branch'
branch_b = data[data['Branch'] == 'B'].reset_index(drop = True).drop('Branch',axis = 1)#dataframe of 'B Branch'
branch_c = data[data['Branch'] == 'C'].reset_index(drop = True).drop('Branch',axis = 1)#dataframe of 'C Branch'
branch_a
```

***We devided data into three dataframes through branchs' column.Dataframe shows that every branchs located in diffirent cities.So we should check again and we need drop unnecessary columns for analyze data better.***

```python
print(sum(branch_a['City'] == 'Yangon')==branch_a['City'].count())#Branch A is located in Yangon city.
print(sum(branch_b['City'] == 'Mandalay')==branch_b['City'].count())#Branch B is located in Mandalay city.
print(sum(branch_c['City'] == 'Naypyitaw')==branch_c['City'].count())#Branch C is located in Naypyitaw.
#Completed checking!
```

```python
branch_a = branch_a.drop('City',axis=1)
branch_b = branch_b.drop('City',axis=1)
branch_c = branch_c.drop('City',axis=1)
```

# **Data Visualization**

```python
data_gender = data.groupby(['Gender','Branch']).count()['Customer type'].unstack(level=0)
fig = px.bar(data_gender, barmode='group',text='value')
fig.update_layout(xaxis_title='Branch', yaxis_title='Count', title='Customer Type by Gender and Branch')
fig.show()
```

*This bar chart shows us which gender is more likely to use branchs.So branch A was always used by males more than females and branch B also by male ,however, branch C was used by females more than males*

```python
customer_type = data.groupby(['Customer type','Branch'])['Invoice ID'].count().unstack(level=0)
fig = px.bar(customer_type, barmode='group',color_discrete_sequence=['green','yellow'],text ='Customer type')
fig.update_layout(xaxis_title='Branchs', yaxis_title='Customers', title='Customer type')
fig.show()
```

```python
fig = px.histogram(data, x='Product line',color = 'Branch')
fig.update_layout(xaxis_title = 'Product lines',title = 'Product lines in market')
fig.show()
```

```python
warnings.simplefilter(action='ignore', category=FutureWarning)
values_category = data.groupby('Product line').sum()['Total']
fig = px.bar(values_category, barmode='group',color_discrete_sequence=['mediumpurple'],text = 'value')
fig.update_layout(xaxis_title='Product lines', yaxis_title='Revenue', title='Revenue of every product lines')
fig.show()
```

```python
marginincome = data['Total'].sum()-data['cogs'].sum()
marginincome
```

```python
sales = data.groupby([data['Datetime'].dt.strftime("%m.%d"),'Branch']).sum()['Total'].reset_index(name='Total')
# sales
sales_jan = sales[sales['Datetime'].str.startswith('01')]
sales_feb = sales[sales['Datetime'].str.startswith('02')]
sales_mar = sales[sales['Datetime'].str.startswith('03')]
fig = px.line(sales_jan, x='Datetime', y='Total', color='Branch', title='January',markers='*')
fig2 = px.line(sales_feb, x='Datetime', y='Total', color='Branch',title='February',markers='*')
fig3 = px.line(sales_mar, x='Datetime', y='Total', color='Branch',title='March',markers='*')
# fig.show()
# fig = px.line(sales, x='Datetime', y='Total',color='Branch')
fig.show()
fig2.show()
fig3.show()
```

**This data of market about 3months work.These 3graphs shows revenue of jan,feb,march**

```python
sales_day = data.groupby([data['Datetime'].dt.strftime("%H:00"),'Branch']).sum()['Total'].reset_index(name='Total')
fig = px.line(sales_day, x='Datetime', y='Total',color = 'Branch',markers='*',title = 'Intraday trading chart')
fig.show()
```

*this graph shows that market works from 10:00 to 20:00*

```python
month_revenue = data.groupby(data['Datetime'].dt.strftime("%B")).sum()['gross income'].reset_index(name='gross income')
# month_revenue
fig = px.pie(month_revenue, values='gross income', names='Datetime',title = 'Difference between month\'s total(in percent)' )
fig.update_traces(textposition='inside', textinfo='percent+label+value')
fig.show()
```

```python
payment_types = data.groupby(data['Payment']).count()['Total'].reset_index(name='Total')
fig = px.pie(payment_types, values='Total', names='Payment',color_discrete_sequence=['lightcyan','cyan','royalblue'])
fig.update_traces(textposition='inside', textinfo='percent+label',title='Payment types in shop(in percents)')
fig.show()
```
